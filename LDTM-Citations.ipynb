```python
def par(liste,typePar="par"):
    result="|".join(liste)
    if typePar=="par":
        return "(%s)"%result
    elif typePar=="nopar":
        return result
    else:
        print "cas non prévu"
        return None

separateurs=[ur'\s+et',ur'\s+and',"\s+\\\\&",","]
regAuteurs=par(listeAuteurs)
regSeparateurs=par(separateurs)
#print regAuteurs, regSeparateurs
```

```python
def proposeKey(auteur,annee,liste=listeRefs):
    keys=set()
    for ref in liste:
        if (auteur in ref[0] and annee in ref[1]) or (auteur=="al."):
            keys.add(ref)
    return list(keys)
```

```python
debugParse=False
def parseCitet(citation):
    target=citation.replace("\n"," ")
    m=re.match(ur"(%s(%s %s)*)('s)? \((\d{4}((,|[ ~];| et ) ?(\d{4}))*)(.*)\)"%(regAuteurs,regSeparateurs,regAuteurs),target)
    if m:
        if debugParse:
            for numG,g in enumerate(m.groups()):
                print numG+1,g
        auteurs=[a.strip(" ") for a in re.split(par(separateurs,"nopar"),m.group(1))]
        annees= [a.strip(" ") for a in re.split(ur" et |~?[,; ]+",m.group(7))]
        head=m.group(6)
        tail=m.group(11)
        return ([(auteurs,annees)],head,tail)
    else:
        if debugParse: print citation
        return [([],[]),"",""]

def parseSubcite(citation):
    m=re.match(ur"(%s(%s %s)*) (\d{4}((,|[ ~];| et ) ?(\d{4}))*)(.*)"%(regAuteurs,regSeparateurs,regAuteurs),citation)
    if m:
        auteurs=[a.strip(" ") for a in re.split(par(separateurs,"nopar"),m.group(1))]
        annees= [a.strip(" ") for a in re.split(ur"(?:,\s*?|[ ~];\s*| et )",m.group(6))]
    return (auteurs,annees)

def parseCitep(citation):
    target=citation.replace(u"\n"," ")
    if debugParse: print target
    publications=""
    noHead=re.match(ur"\(((%s(%s %s)*) (\d{4}((,|[ ~];| et ) ?(\d{4}))*)((,|[ ~]*;| et ) ?(%s(%s %s)*) (\d{4}((,|[ ~];| et ) ?(\d{4}))*))*)(.*)\)"%(regAuteurs,regSeparateurs,regAuteurs,regAuteurs,regSeparateurs,regAuteurs),target)
    if noHead:
        if debugParse:
            for numG,g in enumerate(noHead.groups()):
                print numG+1,g
        subCites=re.findall(ur"((%s(%s %s)*) (\d{4}((,|[ ~];| et ) ?(\d{4}))*))"%(regAuteurs,regSeparateurs,regAuteurs),noHead.group(1))
        publications=[]
        for subCite in subCites:
            publications.append(parseSubcite(subCite[0]))
        head=""
        tail=noHead.group(22)
        return (publications,head,tail)
    else:
        withHead=re.match(ur"\((.+?)((%s(%s %s)*) (\d{4}((,|[ ~];| et ) ?(\d{4}))*)((,|[ ~]*;| et ) ?(%s(%s %s)*) (\d{4}([,;] ?(\d{4}))*))*)(.*)\)"%(regAuteurs,regSeparateurs,regAuteurs,regAuteurs,regSeparateurs,regAuteurs),target)
        if withHead:
            head=withHead.group(1)
            tail=withHead.group(22)
            if debugParse:
                for numG,g in enumerate(withHead.groups()):
                    print numG+1,g
            subCites=re.findall(ur"((%s(%s %s)*) (\d{4}((,|[ ~];| et ) ?(\d{4}))*))"%(regAuteurs,regSeparateurs,regAuteurs),withHead.group(2))
            publications=[]
            for subCite in subCites:
                publications.append(parseSubcite(subCite[0]))
            return (publications,head,tail)
        else:
            print "PB",citation
            return ([("","")],"","")
```

```python
def getNocites(texte):
    mNocite=re.findall(ur"\\nocite\{([^}]*)\}",texte,re.S)
    keys=set()
    for nocite in mNocite:
        for l in nocite.split("\n"):
            for e in l.strip().split(","):
                if e:
                    keys.add(e)
    return sorted([k for k in keys if k!=""])
```
