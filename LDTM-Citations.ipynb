```python
def par(liste):
    return "(%s)"%"|".join(liste)

auteurs=['Fradin', 'Kerleroux', 'Melloni', 'Spencer', 'Thornton',
         'Pustejovsky','Grimshaw'
        ]
separateurs=[ur'\s+et',ur'\s+and',"\s+\\\\&",","]
regAuteurs=par(auteurs)
regSeparateurs=par(separateurs)
print regAuteurs, regSeparateurs
```

```python
def citations(chaine):
    result=[]
    m=re.findall(ur"(%s(%s %s)?('s)? \(\d\d\d\d[^)]*\))"%(regAuteurs,regSeparateurs,regAuteurs),chaine)
    if m:
        for e in m:
            result.append(e[0])
    m=re.findall(ur"(\([^)]*%s(%s %s)? \d\d\d\d[^)]*\))"%(regAuteurs,regSeparateurs,regAuteurs),chaine)
    if m:
        for e in m:
            result.append(e[0])
    return result
```

```python
ligne=ur"Fradin et Kerleroux (2003), ou encore Melloni \& Thornton (1990)"
citations(ligne)
```
