```python
import yaml,re,codecs
```

```python
with codecs.open("/Users/gilles/Downloads/praat.Collection","r",encoding="iso8859-1") as input:
    bruteLines=input.readlines()
```

```python
lines=[l.replace(b"\x00","").strip() for l in bruteLines]
text="\n".join(lines[4:])
```

```python
items=re.split(ur"item \[\d+\]:",text)
```

```python
itemFrames={}
itemNbFrameCandidates={}
for nI,item in enumerate(items):
    frames=re.split("frame \[\d+\]:",item)[1:]
    itemFrames[nI]=frames
    nbFrameCandidates=0
    for frame in frames:
        candidates=re.split("candidate \[\d+\]:",frame)[1:]
        nbFrameCandidates+=len(candidates)
    itemNbFrameCandidates[nI]=nbFrameCandidates
```

```python
nbFrames=0
for kI in itemFrames:
    print kI,len(itemFrames[kI])
    nbFrames+=len(itemFrames[kI])                
```

```python
nbCandidates=0
for kI in itemNbFrameCandidates:
    nbCandidates+=itemNbFrameCandidates[kI]
nbCandidates
```
