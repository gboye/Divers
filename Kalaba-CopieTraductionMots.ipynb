```python
def normalizeLigneP(ligne):
    listeElements=[l[:-1] for l in re.split(r"\\|{|}| ",ligne) if l!=""]
    return " ".join(listeElements)
```

```python
def getTrad(ligne):
    result=ligne.replace("\\begin{preview}","").replace("\\end{preview}","")
    return result
```

```python
nomExemples="Mots.tex"
with codecs.open(serie+nomExemples,"r",encoding="utf8") as input:
    lignesExemples=input.readlines()
```
