```python
import glob,codecs,re
```

```python
fichierNoms=glob.glob("*.txt")
```

```python
questionnaires={}
for n,fichier in enumerate(fichierNoms):
    questionnaires[n]=[]
    with codecs.open(fichier,"r",encoding="utf8")as inFile:
        lines=inFile.readlines()
    debutQuestions=False
    for line in lines:
        if debutQuestions:
            line=line.strip()
            if not line=="":
                questionnaires[n].append(line.strip())
        elif line.startswith("====="):
            debutQuestions=True
    
    
```
