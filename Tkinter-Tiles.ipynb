```python
import Tkinter as tk
import random

dictJetons = {"a": 9, "b": 2, "c": 2, "d": 3, "e": 15, 
         "f": 2, "g": 2, "h": 2, "i": 8, "j": 1, 
         "k": 1, "l": 5, "m": 3, "n": 6, "o": 6, 
         "p": 2, "q": 1, "r": 6, "s": 6, "t": 6, 
         "u": 6, "v": 2, "w": 1, "x": 1, "y": 1, 
         "z": 1, "*": 2}

listJetons=[]
for jeton in dictJetons:
    nombre=dictJetons[jeton]
    listJetons.extend([jeton]*nombre)

numTirage=12
tirage=random.sample(listJetons,numTirage)

class SampleApp(tk.Tk):
    '''Illustrate how to drag items on a Tkinter canvas'''

    def __init__(self, *args, **kwargs):
        tk.Tk.__init__(self, *args, **kwargs)

        # create a canvas
        self.canvas = tk.Canvas(width=400, height=400)
        self.canvas.pack(fill="both", expand=True)

        # this data is used to keep track of an 
        # item being dragged
        self._drag_data = {"x": 0, "y": 0, "item": None}
             
        
        for i,lettre in enumerate(tirage):        
        # create movable objects
            nomToken="token-%d"%i
            self._create_token((random.randrange(400), random.randrange(400)), "blue",nomToken,lettre)
        # add bindings for clicking, dragging and releasing over
        # any object with the "token" tag
            self.canvas.tag_bind(nomToken, "<ButtonPress-1>", self.OnTokenButtonPress)
            self.canvas.tag_bind(nomToken, "<ButtonRelease-1>", self.OnTokenButtonRelease)
            self.canvas.tag_bind(nomToken, "<B1-Motion>", self.OnTokenMotion)

    def _create_token(self, coord, color, tags, lettre):
        '''Create a token at the given coordinate in the given color'''
        (x,y) = coord
        i=self.canvas.create_text(x, y, 
                                fill=color, tags=tags, text=lettre.upper(), 
                                  font=("Helvetica",32))
        
        # normalize rectangle size for letter tile
        largeurCarte=50
        hauteurCarte=50
        (x,y,x1,y1)=self.canvas.bbox(i)
        diffLargeur=largeurCarte-(x1-x)
        if diffLargeur>0:
            x-=diffLargeur/2
            x1+=diffLargeur/2
        diffHauteur=hauteurCarte-(y1-y)
        if diffHauteur>0:
            y-=diffHauteur/2
            y1+=diffHauteur/2
            
        r=self.canvas.create_rectangle(x,y,x1,y1,fill="gray",tags=tags)
        self.canvas.tag_lower(r,i)

    def OnTokenButtonPress(self, event):
        '''Being drag of an object'''
        # record the items and their location
        # get the closest item to the click
        item = self.canvas.find_closest(event.x, event.y)[0]
        # get the tag of the item
        tags = self.canvas.gettags(item)
        # prepare to move all items with the same tag
        self._drag_data["item"] = tags[0]
        self._drag_data["x"] = event.x
        self._drag_data["y"] = event.y

    def OnTokenButtonRelease(self, event):
        '''End drag of an object'''
        # reset the drag information
        self._drag_data["item"] = None
        self._drag_data["x"] = 0
        self._drag_data["y"] = 0

    def OnTokenMotion(self, event):
        '''Handle dragging of an object'''
        # compute how much this object has moved
        delta_x = event.x - self._drag_data["x"]
        delta_y = event.y - self._drag_data["y"]
        # move the object the appropriate amount
        self.canvas.move(self._drag_data["item"], delta_x, delta_y)
        # record the new position
        self._drag_data["x"] = event.x
        self._drag_data["y"] = event.y

if __name__ == "__main__":
    app = SampleApp()
    app.mainloop()
```
