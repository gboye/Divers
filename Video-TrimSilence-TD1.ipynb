```python
import sys
import subprocess
import os
import shutil
from moviepy.editor import VideoFileClip, concatenate_videoclips
```

```python
# https://dev.to/dak425/automatically-trim-silence-from-video-with-ffmpeg-and-python-2kol
rep=u"/Users/gilles/Downloads/"
video_in="TD1"
# Input file path
file_in = rep+video_in+".mp4"
# Output file path
file_out = rep+video_in+"-cut.mp4"
# Silence timestamps
silence_file=rep+"silence-%s.txt"%video_in

ease = 0.0
minimum_duration = 1.0
in_handle = open(silence_file, "r")
video = VideoFileClip(file_in)

def main():
    # number of clips generated
    count = 0
    # start of next clip
    last = 0

#     in_handle = open(silence_file, "r", errors='replace')
    full_duration = video.duration
    print full_duration
    clips = []
    while True:
        line = in_handle.readline()

        if not line:
            break

        end,duration = line.strip().split()
        
        try:
            to = float(end) - float(duration)
        except:
            continue

        start = float(last)
        clip_duration = float(to) - start
        # Clips less than one seconds don't seem to work
        print("Clip Duration: {} seconds".format(clip_duration))

        if clip_duration < minimum_duration:
            continue

        if full_duration - to < minimum_duration:
            continue

        if start > ease:
            start -= ease

        print("Clip {} (Start: {}, End: {})".format(count, start, to))
        clip = video.subclip(start, to)
        clips.append(clip)
        last = end
        count += 1

    if full_duration - float(last) > minimum_duration:
        print("Clip {} (Start: {}, End: {})".format(count, last, 'EOF'))
        clips.append(video.subclip(float(last)-ease))
    return clips
```

```python
# https://stackoverflow.com/questions/54722792/output-video-has-no-sound
def finalize():
    processed_video = concatenate_videoclips(clips)
    processed_video.write_videofile(
        file_out,
        temp_audiofile='temp-audio-%s.m4a'%video_in, remove_temp=True, 
#         fps=60,
#         preset='ultrafast',
        codec='libx264',
        audio_codec="aac"
    )

    in_handle.close()
    video.close()
```
