## Faire la liste des groupes, les étudiants, les fichiers

```python
groupes={}
kanoniks={}
for fichier in fichiers:
    # Dossier-1 A-CLARA CORIDON_63478_assignsubmission_file_
    # Kalabia-01-ALIENOR RIOU_92933_assignsubmission_file_
    # m=re.match(ur"^.*/(Kalabia-\d+)-(.*)_.*_.*_.*_/(.*)$",fichier)
    m=re.match(ur"^.*/(Kanonik-\d+)-(.*)_.*_.*_.*_/(.*)$",fichier)
    if m:
        groupe=m.group(1)
        etudiant=m.group(2)
        fichier=m.group(3)
        if groupe not in groupes:
            groupes[groupe]=set()
        groupes[groupe].add(etudiant)
        if groupe not in kanoniks:
            kanoniks[groupe]=set()
        kanoniks[groupe].add(fichier)
    else:
        print "wrong line", fichier
```

## Copier les fichiers des groupes dans des répertoires par groupe

```python
for kanonik in kanoniks:
    delegue=list(groupes[kanonik])[0]
    for fichier in kanoniks[kanonik]:
        print kanonik,delegue,fichier
        r=re.compile(kanonik+".*"+delegue+".*"+re.escape(fichier))
        lFichiers = filter(r.search, fichiers)
        fSource=lFichiers[0]
        repDestination=rep2+kanonik
#        m=re.match(ur"^(.*)\s?\(.*\)\.(.*)$",fichier)
#        if m:
#            nFichier=m.group(1)+"."+m.group(2)
#            print nFichier
        fDestination=repDestination+"/"+fichier
        print fSource
        print fDestination
        print
        if fSource:
            try:
                shutil.copy2(fSource,fDestination)
            except:
                os.makedirs(repDestination)
                shutil.copy2(fSource,fDestination)
```

## Faire un tableau des groupes

```python
etudiants={}
for groupe in groupes:
    for etudiant in groupes[groupe]:
        etudiants[etudiant]=groupe

dfEtudiants=pd.DataFrame.from_dict(etudiants,orient="index",columns=["groupe"])
print len(dfEtudiants.groupe.unique())
dfEtudiants.sort_values("groupe")
```
