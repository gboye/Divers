```python
import glob,shutil,os
```

```python
base="/Users/gilles/Downloads/2LNSM12-Dépôt pour le travail final sur les kanoniks-45919/"
kanoniks="/Users/gilles/Downloads/24-Kanoniks/"
```

```python
reps=glob.glob(base+"*/")
```

```python
for rep in reps:
    depot=rep.split("/")[-2]
    groupe=depot.split("_")[0]
    dest=kanoniks+groupe+"/"
    fichiers=glob.glob(rep+"*.*")
    for fichier in fichiers:
        print(fichier)
        os.makedirs(os.path.dirname(dest), exist_ok=True)
        shutil.copy(fichier,dest)
```
