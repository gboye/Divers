```python
import codecs,glob
import random as rd
```

```python
nomRepCours="/Users/gilles/ownCloud/Cours/Bordeaux/L1-LinguistiqueGenerale/Kalaba-Project/CoursTemoin/"
nomFichierEntreeXML=nomRepCours+"16-K2-A1-Decoupe-Glose-MC.xml"
```

```python
def lireQuestions(nomFichierEntreeXML):
    with codecs.open(nomFichierEntreeXML,"r",encoding="utf8") as input:
        xmlQuestions=input.readlines()
    questionLines=[]
    for line in xmlQuestions:
        line=line.replace(u"<",u"\n<")
        lines=[l.strip() for l in line.split("\n") if l!=""]
        questionLines.extend(lines) 
    return questionLines
```

```python
def selectionAleatoire(percent=50):
    return rd.randrange(100)<percent
```

```python
fichiersEntreeXML=glob.glob(nomRepCours+"*.xml")
fichiersEntreeXML=[f for f in fichiersEntreeXML if not f.endswith("extrait.xml")]
percentFichier={
                    nomRepCours+"16-K2-A1-Traduction-Radical.xml":20,
                    nomRepCours+"Syntagmes-Ordre-RX.xml":30,
                    nomRepCours+"Traductions-Theme-RX.xml":50,
                    nomRepCours+"Traductions-Version-RX.xml":50
}
for nomFichierEntreeXML in fichiersEntreeXML:
    if nomFichierEntreeXML in percentFichier:
        print "percent",percentFichier[nomFichierEntreeXML]
        ecrireExtrait(nomFichierEntreeXML,percent=percentFichier[nomFichierEntreeXML])
    else:
        ecrireExtrait(nomFichierEntreeXML)
    
```
