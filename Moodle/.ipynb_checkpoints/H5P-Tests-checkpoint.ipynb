# HP5 file manipulation 
la structure des fichiers au sein du .h5p est complexe. c'est une archive ZIP avec beaucoup de fichiers cf. zip.printdir().
- la question est dans content/content.json
- le nom de la question est dans h5p.json

```python
from zipfile import ZipFile
import json
import pandas as pd
```

```python
fSource="/Users/gilles/Downloads/Swann 2.h5p"
```

```python
with ZipFile(fSource, 'r') as zip: 
    zip.printdir()
    contentJson=zip.read("content/content.json")
    h5pJson=zip.read("h5p.json")
```

Il y a une différence entre json.load et json.loads (s pour string)

```python
content=json.loads(contentJson)
h5p=json.loads(h5pJson)
```
