# Génération d'une série d'exercices CLOZE pour l'identification des subordonnées
- un fichier CSV avec les informations
- un accès à ToolsMoodleXML2 pour le packaging

```python
import ToolsMoodleXML2 as TM
import pandas as pd
import codecs
```

```python
csv="/Users/gilles/Downloads/Subordonnees.csv"
xml="/Users/gilles/Downloads/Subordonnees.xml"
df=pd.read_csv(csv,sep="\t",encoding="utf8")
df.head()
```

```python
questions=TM.XMLSeries("Subordination/Identification/SAC")
```

```python
for i,r in df.iterrows():
    num="%02d"%int(r["index"])
    sub=r["subordonnant"]
    start=r["start"]#.replace("/","~=")
    end=r["end"]#.replace("/","~=")
    cat=r["type"]
    gouv=r["gouv"].replace("X","PHRASE")
    titre=u"%s %s : %s [%s—%s] => %s"%(num,sub,cat,start,end,gouv)
    print titre
    sac="{1:SAC:=%s}"
    sacStart=sac%start.replace("/","~=")
    sacEnd=sac%end.replace("/","~=")
    sacCat=sac%cat
    sacGouv=sac%gouv.replace("/","~=")
    consigne=u"Pour la subordonnée %s<sub>%s</sub> du <a href='https://e-campus.u-bordeaux-montaigne.fr/mod/resource/view.php?id=25245' target='_blank'>document</a>, donnez le premier et le dernier mot de la construction, son type et son rattachement pour les relatives/complétives/questions indirectes (ou la catégorie remplacée pour les constructions sans antécédent NP/PP/PHRASE)<br>Début : %s<br>Fin : %s<br>Type : %s<br>Référent : %s"%(sub,num,sacStart,sacEnd,sacCat,sacGouv)
    question=TM.ClozeExercice(titre,consigne,penalty=0.2)
    questions.addExercice(question)
```

```python
xmlOutput=TM.MoodleXML()
xmlOutput.addQuiz(questions.getSeries())
with codecs.open(xml, 'w',encoding="utf8") as output:
    output.write(xmlOutput.getXML())
```
