```python
import glob,re,filecmp,os
import itertools as it
```

```python
repertoire=u"/Users/gilles/Downloads/5LNSE12-Dossier 1  Présentation de la conjugaison-958/"
devoir="Conjugaison1"
#repertoire=u"/Users/gilles/Downloads/MNT2U513-MNN2M21-Devoir/"
fichiers=glob.glob(repertoire+u"*submission*")
```

```python
patronGroupe=ur"^[^-]+-([^_]+)_.*$"
patronIndividuel=ur"^([^_]+)_.*$"
fichierEtudiants={}
for nFichier,fichier in enumerate(fichiers):
    nom=fichier.split(u"/")[-1]
    m=re.match(patronGroupe,nom)
    if m:
        fichierEtudiants[fichier]=(m.group(1),nFichier)
fichierEtudiants        
        
```

```python
etudiantsGroupe={}
for f1,f2 in it.combinations(fichiers,2):
    if filecmp.cmp(f1,f2):
        nom1,num1=fichierEtudiants[f1]
        nom2,num2=fichierEtudiants[f2]
        if not nom1 in etudiantsGroupe:
            etudiantsGroupe[nom1]=num1
        else:
            num1=etudiantsGroupe[nom1]
        etudiantsGroupe[nom2]=num1
print etudiantsGroupe
```

```python
groupeEtudiants = {}
for k, v in etudiantsGroupe.iteritems():
    groupeEtudiants[v] = groupeEtudiants.get(v, [])
    groupeEtudiants[v].append(k)
groupeEtudiants
```

```python
repertoire=u"/Users/gilles/Downloads/LNS6E812-Dépot pour devoir sur table-86643/"
#repertoire=u"/Users/gilles/Downloads/MNT2U513-MNN2M21-Devoir/"
fichiers=glob.glob(repertoire+u"*submission*")
```

```python
nomsEtudiant=[]
for f in fichiers:
    m=re.match(ur"(.*/([^/]+))_\d+_assignsubmission_file_(.*)",f)
    if m:
        fPartie1=m.group(1)
        etu=m.group(2)
        fPartie2=m.group(3)
        os.rename(f,fPartie1+"-"+fPartie2)
        nomsEtudiant.append(etu)
nomsEtudiant
```
