```python
import codecs,glob,re
import random as rd
import itertools as it
```

```python
nbLiens=3
rd.seed(2464)
nomRepCours="/Users/gilles/Downloads/"
nomFichierEntreeXML=nomRepCours+"quiz-LNS2U2-DSTM-20200514-1221.xml"
nomFichierLiens="/Users/gilles/pCloud Drive/FOD/GB/L1-Morphologie/LiensMorphoL1.txt"
```

```python
with codecs.open(nomFichierEntreeXML,"r",encoding="utf8") as input:
    xmlQuestions=[l.strip() for l in input.readlines()]
```

```python
def getTemplate(xmlLines,qType="essay"):
    header=[]
    questionTemplate=[]
    trailer=[]
    boolTemplate=False
    boolTrailer=False
    for xmlLine in xmlLines:
        m=re.search(ur'<question type="essay">',xmlLine)
        if m:
            questionTemplate.append(xmlLine)
            boolTemplate=True
        elif ur"</question>" in xmlLine and boolTemplate:
            questionTemplate.append(xmlLine)
            boolTemplate=False
            boolTrailer=True
        elif boolTemplate:
            questionTemplate.append(xmlLine)
        elif boolTrailer:
            trailer.append(xmlLine)
        else:
            header.append(xmlLine)
    return header, questionTemplate, trailer
```

```python
header,templateLines,trailer=getTemplate(xmlQuestions)
```

```python
with codecs.open(nomFichierLiens,"r",encoding="utf8") as input:
    liens=[l.strip() for l in input.readlines()]
liens
```

```python
def newQuestion(template,i,j):
    numLien=nbAnalyses*i+j
    result=[]
    for line in template:
        if "DST Pxx-Axx" in line:
            result.append(line.replace("DST Pxx-Axx","DST P%d-A%d"%(i+1,j+1)))
        elif "http://urlxxx" in line:
            result.append(line.replace("http://urlxxx",liens[numLien]))
        else:
            result.append(line)
    return result
```

```python
def newQuestions(templateLines):
    result=[]
    for i in range(nbPhenomenes):
        for j in range(nbAnalyses):
            question=newQuestion(templateLines,i,j)
            result.extend(question)
    return result
```

```python
nbPhenomenes=3
nbAnalyses=3
sortieXML=header+newQuestions(templateLines)+trailer
print "\n".join(sortieXML)
```
