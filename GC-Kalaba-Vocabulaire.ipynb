```python
with codecs.open(serie+"Phrases-Clozes.txt","r",encoding="utf8") as input:
    clozes=[l.strip() for l in input.readlines() if l.strip()!=u""]
```

```python
vocabulaire=set()
lexique={}
for l in clozes:
    if not l.startswith(u"#"):
        print l
        lEls=l.split(";")
        citation=lEls[1]
        cat=lEls[2]
        radical=lEls[3]
        traits=lEls[7].split(u",")
        lexique[radical]=citation
        if cat=="NOM":
            cf="%s:%s"%(traits[1].split("=")[1],traits[2].split("=")[1])
        else:
            cf=traits[1].split("=")[1]
        vocabulaire.add((cat,cf, radical,citation))
vocabulaire
```

```python
with codecs.open(serie+"Stems.yaml","r",encoding="utf8") as inFile:
    stems=yaml.load(inFile)
```

```python
from random import shuffle

def chunkify(lst,n):
    return [lst[i::n] for i in xrange(n)]

lignes=[u"%s: [%s]"%(k,",".join(v)) for k,v in stems["Items"].items()]
shuffle(lignes)
for l in lignes:
    print l
```

```python
cf=chunkify(lignes,16)
for i in range(2):
    nCF=i+1
    print "\tN%d:"%nCF
    for j in range(2):
        sCF=j+1
        print "\t\tN%d%d:"%(nCF,sCF)
        for n,k in enumerate("ABCD"):
            print "\t\t\t%s:"%k
            for l in cf[8*i+4*j+n]:
                print "\t\t\t\t%s"%l
```
