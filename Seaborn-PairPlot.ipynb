```python
# library & dataset
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('penguins')
df
```

```python
# Basic correlogram
sns.pairplot(df, kind="reg",hue="species",height=5,markers=["o", "s", "D"])
plt.show()
```

```python
df["cat"]=df.species+":"+df.island
sns.pairplot(df, kind="reg",hue="cat",height=5)
plt.show()
```
