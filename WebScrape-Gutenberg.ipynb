## Préparation du traitement des pages HTML

- définir la page d'accueil de Wiktionary

```python
import mechanicalsoup
import bs4
import re
baseURL="https://www.gutenberg.org"
```

```python
browser = mechanicalsoup.Browser()
```

```python
liste=["œuvres+complètes+flaubert"]
```

```python
for auteur in liste:
    pageIndex=browser.get(baseURL+"/ebooks/search/?query=%s"%auteur)
```

```python
def splitLivre(text):
    try:
        text=re.split("\*\*\* START OF THE PROJECT GUTENBERG EBOOK .* \*\*\*",text)[1]
    except:
        print("pas de découpe sur ponctuation")
    try:
        text=text.split("La ponctuation a pu faire l'objet de quelques corrections mineures.")[1]
    except:
        print("pas de découpe sur ponctuation")
    try:
        text=text.split("La liste des modifications se trouve à la fin du texte.")[1]
    except:
        print("pas de découpe sur modifications") 
    return text
```

```python
livres=pageIndex.soup.findAll(class_="booklink")
for livre in livres:
    titre=livre.find(class_="title").text
    print(titre)
    lien=livre.find(class_="link")["href"]
    text=browser.get(baseURL+lien+".txt.utf-8").text
    text=splitLivre(text)
    fText=titre+".txt"
    with open(fText,"w") as outFile:
        outFile.write(text)
```
