```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame({'Col A': ['syntaxe', 'pragmatique', 'discours', 'vocabulaire', 'registres'],
                   'Enfant1': [2, 2, 2, 3, 1],
                  'Enfant2': [2, 3, 2, 1, 3]})
fig = plt.figure(figsize=(5,5))



ax = fig.add_subplot(111, projection="polar")


# theta has 5 different angles, and the first one repeated
theta = np.arange(len(df) + 1) / float(len(df)) * 2 * np.pi
# values has the 5 values from 'Col B', with the first element repeated
values = df['Enfant1'].values
values = np.append(values, values[0])

# draw the polygon and the mark the points for each angle/value combination
l1, = ax.plot(theta, values, color="blue", marker="o", label="Enfant1")
plt.xticks(theta[:-1], df['Col A'], color='grey', size=12)
# ax.tick_params(pad=10) # to increase the distance of the labels to the plot
# fill the area of the polygon with green and some transparency
ax.fill(theta, values, 'blue', alpha=0.5)


values = df['Enfant2'].values
values = np.append(values, values[0])

# draw the polygon and the mark the points for each angle/value combination
l2, = ax.plot(theta, values, color="green", marker="o", label="Enfant2")
plt.xticks(theta[:-1], df['Col A'], color='grey', size=12)
ax.tick_params(pad=12) # to increase the distance of the labels to the plot
# fill the area of the polygon with green and some transparency
ax.fill(theta, values, 'green', alpha=0.5)

ax.set_rticks([1, 2, 3])
# ax.set_rlabel_position(22.5)
ax.set_rmax(3)

circle = plt.Circle((0, 0), 2, transform=ax.transData._b,alpha=0.5,linewidth=10,linestyle="-",fill=False)
ax.add_artist(circle)


# plt.legend(loc='upper left') # shows the legend, using the label of the line plot (useful when there is more than 1 polygon)
plt.legend(bbox_to_anchor=(0.95, 1), loc=2, borderaxespad=0.)
plt.title("Profils", y=1.1)
plt.show()
```
