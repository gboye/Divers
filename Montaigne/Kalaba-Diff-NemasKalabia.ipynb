```python
import re, pyperclip, codecs
```

```python
with codecs.open("/Users/gilles/Downloads/Exemples-Nemas.tex",encoding="utf8") as inFile:
    nemas=inFile.readlines()
```

```python
with codecs.open("/Users/gilles/Downloads/Exemples-Kalabia.tex",encoding="utf8") as inFile:
    kalabia=inFile.readlines()
```

```python
n=1
tNemas=[]
for l in nemas:
    m=re.match(ur"^[^%#\\]",l)
    if m:
        ligne="\\ex %s \\xe"%(l.strip())
        tNemas.append(ligne)
        n+=1
tCopy=u"\n".join(tNemas)
pyperclip.copy(tCopy)
```

```python
n=1
tKalabia=[]
for l in kalabia:
    m=re.match(ur"^[^%#\\]",l)
    if m:
        ligne="\\ex %s \\xe"%(l.strip())
        tKalabia.append(ligne)
        n+=1
tCopy=u"\n".join(tKalabia)
pyperclip.copy(tCopy)
```
