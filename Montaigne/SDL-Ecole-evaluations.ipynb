```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from io import StringIO
import seaborn as sns
```

```python
tableau='''Dimensions	Camille	Jeanne	3	4	5	Gaëlle	Léo	Coleen	9	Maureen	Maewenn	12	Elise	Laëtitia	Eléna	16	Sascha
Orthographe	M20/S16	20/18	20/18		20/18		20/16	20/16		20/16	20/17	20/16	20/15	20/17		20/16	20/16
Syntaxe		19/15				M:18/S:16	14/14	16/14							17 / 15		15/15
Cohérence		19/16	20/14		20/14	M:18/S:17	20/15	18/15		20/17	20/16	20/14				20/15	20/20
Connecteurs	M14/S16	18/14					16/14						17/17	14/17			
Conjugaison		19/17	20/17	20/14	18/16	M:18/S:15	20/17	20/16	20/16	20/15	20/16	20/15			18 / 14	20/15	20/17
Longueur des énoncés	M17/S17																
Graphie								15/17		18/19	19/17		17/19	18/19	19/19		14/18
Niveau de langue			12/18		12/19												
Variété du vocabulaire 			12/18		15/19					19/14	18/15	19/17				18/16	
Organisation du discours	M20/S20					M:18/S:18							17/18	17/18	18 / 16		
Homophones				20/20					20/20								
Accords				20/16					20/17								
Ponctuation				18/13		M:18/S:15			18/12	19/14	19/14	19/11				18/11	
Lisibilité	M17/S19			20/20					20/20				17/19	17/18			'''
```

```python
df=pd.read_csv(StringIO(tableau),sep="\t")
df
```

```python
df.Camille=df.Camille.str.replace("M","")
df.Camille=df.Camille.str.replace("S","")
df["Gaëlle"]=df["Gaëlle"].str.replace("M:","")
df["Gaëlle"]=df["Gaëlle"].str.replace("S:","")
df=df.set_index("Dimensions")
noms=df.columns.tolist()
df.head()
```

```python
cols=df.columns.tolist()
for c in cols:
    df[[c+"-M",c+"-S"]]=df[c].str.split('/', expand=True)
    df[c+"-M"]=df[c+"-M"].astype("Int64")
    df[c+"-S"]=df[c+"-S"].astype("Int64")
```

```python
colsM=[c for c in df.columns if "-M" in c]
colsS=[c for c in df.columns if "-S" in c]
```

```python
colsM
```

```python
notes={}
for nom in noms:
    dist={}
    cM=nom+"-M"
    cS=nom+"-S"
    rM=df[cM].T.astype("Int64")
    dM=rM.value_counts().to_dict()
    rS=df[cS].T.astype("Int64")
    dS=rS.value_counts().to_dict()
    for i in range(1,21):
        dist[i]=0
        if i in dM:
            dist[i]+=dM[i]
        if i in dS:
            dist[i]+=dS[i]
    notes[nom]=(rM.mean(),rM.std(),rS.mean(),rS.std(),dist)
```

```python
dfNotes=pd.DataFrame(notes,index=["meanM","stdM","meanS","stdS","dist"]).T
dfDist=pd.json_normalize(dfNotes.dist)
```

```python
bright=sns.color_palette("bright",16)
```

```python
fig,axes=plt.subplots(figsize=(8,6),nrows=4, ncols=4)
fig.tight_layout()
for n in range(16):
    i=n//4
    j=n%4
    dfDist.iloc[n].plot(ax=axes[i,j],kind="bar",color=bright[n])
    axes[i,j].set_ylim([0, 6])
    axes[i,j].set_xlim([9, 20])
    axes[i,j].tick_params(axis='both',labelsize=6)
```

```python
# fig,axes=plt.subplots(figsize=(6,6),nrows=2, ncols=3)
fig, axes = plt.subplot_mosaic("ABCCC;DEFFF",figsize=(10,6))
pastel=sns.color_palette("pastel")
labelsize=8

dfA=dfNotes["meanM"].groupby(pd.cut(dfNotes["meanM"],np.arange(14,21,1)),observed=False).count()
dfA.plot(ax=axes["A"],kind="bar",color=pastel[0])
axes["A"].tick_params(axis='both',labelsize=labelsize)
axes["A"].set_ylim([0, 10.5])

dfB=dfNotes["stdM"].groupby(pd.cut(dfNotes["stdM"],np.arange(0,5,1)),observed=False).count()
dfB.plot(ax=axes["B"],kind="bar",color=pastel[0])
axes["B"].tick_params(axis='both',labelsize=labelsize)
axes["B"].set_ylim([0, 10.5])

dfC=df[colsM].T.astype("Int64").std()
dfC.plot(ax=axes["C"],color=pastel[0])
axes["C"].tick_params(axis='both',labelsize=labelsize)
axes["C"].set_ylim([0, 3])

dfD=dfNotes["meanS"].groupby(pd.cut(dfNotes["meanS"],np.arange(14,21,1)),observed=False).count()
dfD.plot(ax=axes["D"],kind="bar",color=pastel[1])
axes["D"].tick_params(axis='both',labelsize=labelsize)
axes["D"].set_ylim([0, 10.5])

dfE=dfNotes["stdS"].groupby(pd.cut(dfNotes["stdS"],np.arange(0,5,1)),observed=False).count()
dfE.plot(ax=axes["E"],kind="bar",color=pastel[1])
axes["E"].tick_params(axis='both',labelsize=labelsize)
axes["E"].set_ylim([0, 10.5])

dfF=df[colsS].T.astype("Int64").std()
dfF.plot(ax=axes["F"],color=pastel[1])
axes["F"].tick_params(axis='both',labelsize=labelsize)
axes["F"].set_ylim([0, 3])

x=df.index
xMean="14-15 15-16 16-17 17-18 18-19 19-20".split(" ")
xStd="0-1 1-2 2-3 3-4".split(" ")
plt.sca(axes["A"])
plt.xticks(range(len(xMean)),xMean,rotation=45,ha='right')
plt.title("Moyennes Marie")
plt.xlabel("")
plt.ylabel("Nombre d'occurrences")
plt.sca(axes["B"])
plt.xticks(range(len(xStd)),xStd,rotation=45,ha='right')
plt.title("Écarts Marie")
plt.xlabel("")
plt.sca(axes["C"])
plt.xticks(range(len(x)),x,rotation=45,ha='right')
plt.title("Écarts Marie par critère")
plt.xlabel("")
plt.ylabel("Écart type\nentre les évaluations")
plt.sca(axes["D"])
plt.xticks(range(len(xMean)),xMean,rotation=45,ha='right')
plt.title("Moyennes Samantha")
plt.xlabel("")
plt.ylabel("Nombre d'occurrences")
plt.sca(axes["E"])
plt.xticks(range(len(xStd)),xStd,rotation=45,ha='right')
plt.title("Écarts Samantha")
plt.xlabel("")
plt.sca(axes["F"])
plt.xticks(range(len(x)),x,rotation=45,ha='right')
plt.title("Écarts Samantha par critère")
plt.xlabel("")
plt.ylabel("Écart type\nentre les évaluations")
fig.tight_layout()
```
