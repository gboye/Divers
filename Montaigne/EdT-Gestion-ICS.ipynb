```python
import datetime,re
import icalendar as ical
from pathlib import Path
from pytz import UTC # timezone
```

```python
ics="/Users/gilles/Downloads/Export.ics"
annee="2025"
mois="09"
sDays=-15
```

```python
ics_path = Path(ics)
decalage=371
calendar = ical.Calendar.from_ical(ics_path.read_bytes())
# print(calendar)
for element in calendar.walk():
    if element.name=="VEVENT":
        print(element["SUMMARY"],(element.get("DTSTART").dt+datetime.timedelta(days=decalage)),)
        
```

```python
def changeDate(chaine):
    tag,dt=chaine.split(":")
    m=re.match(r"(\d\d\d\d)(\d\d)(\d\d)(.*)",dt)
    if m:
        date=annee+mois+"%02d"%(int(m.group(3))+sDays)+m.group(4)
    return tag+":"+date+"\n"
```

```python
with open(ics) as inFile:
    lines=inFile.readlines()

outLines=[]
for line in lines:
    if "DTSTART" in line or "DTEND" in line:
        outLines.append(changeDate(line))
    else:
        outLines.append(line)

with open(ics.replace(".ics","-mod.ics"),"w") as outFile:
    for outLine in outLines:
        outFile.write(outLine)
```
