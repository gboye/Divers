```python
def listFichiersK(repertoire):
    result={}
    for rep,subReps,fichiers in os.walk(repertoire):
        if subReps==[]:
            groupe=rep.split("/")[-1]
            result[groupe]={}
            for fichier in fichiersAnalyse:
                groupeAnalyse=groupe+fichier
                analyse=fichier.split(".")[0][1:]
                if groupeAnalyse in fichiers:
                    result[groupe][analyse]=groupeAnalyse
            groupeTD=groupe+fichierTD
            if groupeTD in fichiers:
                result[groupe]["TD"]=groupeTD
            for fichier in fichiersTransfert:
                if fichier in fichiers:
                    result[groupe][fichier]=fichier
    return result
```
