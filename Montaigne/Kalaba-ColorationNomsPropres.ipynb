```python
import re,yaml,codecs,os
from os.path import expanduser

def ding():
    os.system('afplay /System/Library/Sounds/Submarine.aiff')
```

```python
home = expanduser("~")
```

```python
nKalaba="K5"
annee="23"
repertoire=home+"/ownCloud/Cours/Bordeaux/L1-LinguistiqueGenerale/00-ProjetKalaba/%s-%s"%(annee,nKalaba)
serie=repertoire+"/"
```

```python
nomDeclarations="Declarations.tex"
with codecs.open(serie+nomDeclarations,"r",encoding="utf8") as input:
    lignesDeclarations=[l.strip() for l in input.readlines()]
```

```python
personnages="Mahira Violette Kaleb Nicole Katisha Nabil Agathos Elphaba Freja Clemencia Hoderi".split(" ")
persoRegex="|".join(personnages)
newDeclarations=[]
for ligne in lignesDeclarations:
    m=re.match(ur"^(.*(%s).*)grapho(\{.*)"%persoRegex,ligne)
    if m:
        perso=m.group(2)
        newDeclarations.append(ligne.replace(ur"\grapho{",ur"\grapho%s{"%perso))
    else:
        newDeclarations.append(ligne)
```

```python
for d in newDeclarations:
    print d
```

```python
with codecs.open(serie+nomDeclarations.replace(".tex","-Texte.tex"),"w",encoding="utf8") as outfile:
    for ligne in newDeclarations:
        outfile.write(ligne+"\n")
```
