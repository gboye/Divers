```python
def getGrimm(code):
    fName=repertoire+"Grimm-%s.text"%code
    with codecs.open(fName,"r",encoding="utf8") as infile:
        texte="".join(infile.readlines())
    return texte
```

```python
def getTale(code):
    table={}
    text=getGrimm(code)
    tales=re.findall(ur"(\d{3})\s+—(.*?)(?=\d{3}\s+—)",text,re.DOTALL)
    print code,len(tales)
    for tale in tales:
        table[tale[0]]=tale[1]
    return table
    
```

```python
filtrePonct=u":,;()\t- '`…–‚«»"+'"'

def scramble(code,num,filtre=True):
    result=num+u" — "
    text=bs4.BeautifulSoup(tables[code[:2]][num]).text
    if filtre:
        for lettreBrute in text:
            if not lettreBrute in filtrePonct:
                result+=lettreBrute.lower()
                if lettreBrute in "?.!":
                    result+=" "
    else:
        result+=text
    return result
```

```python
def diphtonguer(chaine,langue="FR"):
    result=chaine
    diphtongues=diphtonguaisons[langue]
    for d in diphtongues:
        result=result.replace(d,diphtongues[d])
    return result

def trans(chaine,langue="FR"):
    result=diphtonguer(chaine,langue)
    toC,toV=transTable(langue)
    result=result.translate(toC)
    result=result.translate(toV)
    return result
```
