```python
import pandas as pd
```

```python
fPostes="/Users/gilles/Downloads/ListePostePubliesAnnuelle-GB.xlsx"
dfPostes=pd.read_excel(fPostes)
```

```python
for col in dfPostes.columns[dfPostes.isna().any()].tolist():
    if dfPostes[col].dtype == 'float':
        dfPostes[col] = dfPostes[col].astype('Int64')
dfPostes
```

```python
bSection7=(dfPostes["Section"]==7) | (dfPostes["Section 2"]==7) | (dfPostes["Section 3"]==7) 
dfSection7=dfPostes[bSection7]
dfSection7[["Etablissement","Corps","Section","Section 2","Section 3","Profil","Localisation"]]
dfS7Mcf=dfSection7[dfSection7.Corps=="MCF"][["Etablissement","Corps","Section","Section 2","Section 3","Profil","Localisation"]]
dfS7Mcf.sort_values(["Etablissement","Section","Section 2","Section 3"]).to_csv(fPostes.replace(".xlsx","-MCF.csv"),encoding="utf8",sep="\t")
```
