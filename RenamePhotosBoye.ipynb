```python
import os,glob,re
from PIL import Image
```

From https://stackoverflow.com/questions/23064549/get-date-and-time-when-photo-was-taken-from-exif-data-using-pil

```python
rep="/Volumes/MiniExt/"
photos=glob.glob(rep+"*.JPG")+glob.glob(rep+"*.jpg")

for i,photo in enumerate(photos[:]):
    if i%100==0: print("\n%d\n"%(i))
    nOrigin=photo.rsplit("/",1)[1]
    m=re.search(r"^\d{6}-",nOrigin)
    if not m:
        try:
            im=Image.open(photo)
        except:
            print(photo)
        exif=im._getexif()
        if exif:
            date=exif.get(36867)
            if date:
                annee,mois,jour=date.split(" ")[0].split(":")
                annee=annee[2:]
                nNew="%s%s%s"%(annee,mois,jour)+"-"+nOrigin
            else:
                nNew="999999"+"-"+nOrigin
            # print(nNew)
            os.rename(rep+nOrigin,rep+nNew)
        else:
            nNew="999999"+"-"+nOrigin
            # print(nNew)
            os.rename(rep+nOrigin,rep+nNew)
    else:
        print(nOrigin,end=",")
```
