```python
import random as rd
import pandas as pd
```

```python
def shuffleChoix(liste):
    result=liste[:]
    rd.shuffle(result)
    return result
```

```python
groupes={"GroupeA":25,"GroupeB":25,"GroupeC":20,"GroupeD":10}
choix=groupes.keys()
etudiants=pd.DataFrame(data=[shuffleChoix(choix) for i in range(75)])
etudiants
```

```python
effectifsGroupes=dict(groupes)
affectations={}
for index,etudiant in etudiants.iterrows():
    if effectifsGroupes[etudiant[0]]>0:
#        print index,etudiant[0]
        affectations[index]=etudiant[0]
        effectifsGroupes[etudiant[0]]-=1
    elif effectifsGroupes[etudiant[1]]>0:
#        print index,etudiant[1]
        affectations[index]=etudiant[1]
        effectifsGroupes[etudiant[1]]-=1
    elif effectifsGroupes[etudiant[2]]>0:
#        print index,etudiant[2]
        affectations[index]=etudiant[2]
        effectifsGroupes[etudiant[2]]-=1
    elif effectifsGroupes[etudiant[3]]>0:
#        print index,etudiant[3]
        affectations[index]=etudiant[3]
        effectifsGroupes[etudiant[3]]-=1
    else:
        print "pas d'affectation pour "+index
#affectations
```

```python
etudiants["affectations"]=affectations.values()
```
