```python
import requests
from bs4 import BeautifulSoup
```

```python
params = {'ctl00$cphMainContent$m_ctrlSearchEngine$m_ctrlRechercheForm$m_txtSearch': 'pommes'}
r = requests.post("http://www.marmiton.org/recettes/recherche.aspx?aqt=pommes", data=params)
#print(r.text)
```

```python
q = BeautifulSoup(r.text)
```

```python
links = q.findAll("a")
for li in links:
    if 'href' in li.attrs:
        print(li['href'])
```
