## Word ngrams

```python
lxaCorpus.wordlist()
lxaCorpus.word_unigram_counter()
```

```python
seuilUnigram=1000
dUnigram=lxaCorpus.word_unigram_counter()
pUnigram=sorted(dUnigram.items(), key=lambda kv: (kv[1], kv[0]), reverse=True)
print "mots de fréquence supérieure à %d"%seuilUnigram
for w,f in pUnigram:
    if f>seuilUnigram:
        print w,f
    else:
        break
```

```python
lxaCorpus.word_bigram_counter()
lxaCorpus.word_trigram_counter()
```

```python
trigrams=lxaCorpus.word_trigram_counter()
for trigram, count_ in sorted(trigrams.items(), key=lambda x: x[1], reverse=True):
    print(trigram, count_)

    if count_ < 100:
        break
```
