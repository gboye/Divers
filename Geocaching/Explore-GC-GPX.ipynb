# Récupérer les caches d'un GPX
Les coordonnées fournies sont les corrigées

```python
from lxml import etree
import pandas as pd
import math
debug=False
```

```python
parser = etree.XMLParser(remove_blank_text=True)
```

```python
#nomLoc="/Users/gilles/Downloads/18303005_ComtéduLabourd/18303005_ComtÇ du Labourd.gpx"
#wptChildren=["name","urlname"]
nomLoc="/Users/gilles/Downloads/caches.gpx"
wptChildren=["name","coord","link"]
```

```python
def localName(element):
    return etree.QName(element).localname
```

```python
def Dec2Min(coord):
    degreSign=""
    decDegres,intDegres=math.modf(coord)
    minutes=abs(decDegres)*60
    return "%d%s %5.3f"%(intDegres,degreSign,minutes)
```

```python
def CoordDec2Min(lat,lon):
    sep=" "
    coordLat=Dec2Min(lat)
    coordLon=Dec2Min(lon)
    if coordLat.startswith("-"):
        coordLat="S"+coordLat[1:]
    else:
        coordLat="N"+coordLat
    if coordLon.startswith("-"):
        coordLon="W"+coordLon[1:]
    else:
        coordLon="E"+coordLon
    return [coordLat,sep,coordLon]
```

```python
def dictLatLon2Coord(dictLatLon):
    return CoordDec2Min(float(dictLatLon["lat"]),float(dictLatLon["lon"]))
```

```python
tree = etree.parse(nomLoc)
root = tree.getroot()
print root.nsmap
```

```python
gcCodes=[]
for wp in root.findall("{%s}wpt"%root.nsmap[None]):
    if debug: print localName(wp), 
    print "%s %s %s"%(dictLatLon2Coord(wp.attrib)[0],dictLatLon2Coord(wp.attrib)[1],dictLatLon2Coord(wp.attrib)[2])
    for element in wp:
        if localName(element) in wptChildren:
            print "\t",localName(element),element.text
            gcCodes.append([element.text]+dictLatLon2Coord(wp.attrib))
            if len(element.attrib)>0:
                for attrib in element.attrib:
                    print "\t - %s : %s,"%(attrib,element.attrib[attrib])
                print
            if len(element) >0:
                cache=element
                for node in cache:
                    if localName(node) in wptChildren:
                        print "\t\t",localName(node),
                        if node.text:
                            print node.text.replace("\n"," "),
                        print
```

```python
dfCaches=pd.DataFrame.from_records(gcCodes)
dfCaches.loc[:,[0,1,3]].to_csv("Caches.csv",sep="\t",index=False,header=None)
```
