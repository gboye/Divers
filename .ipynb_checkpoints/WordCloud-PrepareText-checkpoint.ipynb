```python
import codecs,re
from stop_words import get_stop_words
```

```python
exceptions=True
exceptedWords=[]
exceptedWords=exceptedWords+u"us celle-ci without toutes toute url http quelques quelque xx get red green orange little temps".split(" ")
exceptedWords=exceptedWords+u"trade après see bien tour got one two vers plus aujourd hui back way fin end".split(" ")
exceptedWords=exceptedWords+u"jusqu utc part even just petit time road place bit déjà fini site size nice".split(" ")
exceptedWords=exceptedWords+u"right day next côté still around blue petite first monde vue side met caches".split(" ")

enStops = get_stop_words('en')
frStops = get_stop_words('fr')
spStops = get_stop_words('spanish')
inStops = get_stop_words('indonesian')
stops=enStops+frStops+spStops+inStops
```

```python
with codecs.open("test/logsTexte.txt","r",encoding='utf8') as input:
    texte=input.readlines()
```

```python
lines=[]
wordsCount={}
for line in texte:
    words=re.split(ur"['\s(){}[\]#/*\"“”‘’,;:.-=]+",line)
    words=[w.lower() for w in words if (not w.lower() in stops) and (w!="")]
    for word in words:
        if not word in wordsCount:
            wordsCount[word]=0
        wordsCount[word]+=1
    lines.append(" ".join(words))
print len(lines)
```

```python
with codecs.open("test/logsTexteFiltered.txt","w",encoding='utf8') as output:
    output.writelines(lines[:5000])
    output.writelines(lines[5000:10000])
    output.writelines(lines[10000:15000])
    output.writelines(lines[15000:])
```

```python
filteredWords=[]
for word in wordsCount:
    if (wordsCount[word]>250) and (len(word)>1) and (not word.startswith("-")) and not ((word in exceptedWords) and (exceptions)):
#        print u"%s:%d"%(word,wordsCount[word])
        filteredWords.append(word)
filteredWords    
```

```python
filteredLines=[]
for line in texte:
    words=re.split(ur"['\s(){}[\]#/*\"“”‘’,;:.-=]+",line)
    words=[w for w in words if w.lower() in filteredWords]
    if words:
        filteredLines.append(" ".join(words)+u"\n")
```

```python
with codecs.open("test/logsTexteFilteredWords.txt","w",encoding='utf8') as output:
    output.writelines(filteredLines[:5000])
    output.writelines(filteredLines[5000:10000])
    output.writelines(filteredLines[10000:15000])
    output.writelines(filteredLines[15000:])
```
