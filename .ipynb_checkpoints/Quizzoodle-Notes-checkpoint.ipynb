```python
def assemblerNoms(nom,sep=";"):
    noms=nom.replace(" ",sep).replace("-",sep).replace("'",sep).split(sep)
    result=[]
    for ordre in it.permutations(noms):
        result.append("".join(ordre))
    return result
```

```python
'''https://stackoverflow.com/a/35491326/3986209'''
def invert_dict(d): 
    inverse = dict() 
    for key in d: 
        # Go through the list that is saved in the dict:
        for item in d[key]:
            # Check if in the inverted dict the key exists
            if item not in inverse: 
                # If not create a new list
                inverse[item] = [key] 
            else: 
                inverse[item].append(key) 
    return inverse
```

```python
invAlias=invert_dict(alias)
correspondances={"alias2etu":alias,"etu2alias":invAlias}
```

```python
with codecs.open(fAliasQDLE, 'w', encoding="utf8") as outfile:
    yaml.safe_dump(correspondances, outfile, default_flow_style=False)
```
