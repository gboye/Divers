```python
import pandas as pd
import numpy as np
```

```python
notesNom="/Users/gilles/Downloads/LNS1M41 Notes-Majeure.csv"
dfNotes=pd.read_csv(notesNom,sep="\t",encoding="utf8")
```

```python
dfNotes.set_index([u'Nom', u'Prénom'],inplace=True)
```

#### sélection des colonnes
- notesCC pour le contrôle continu
- notesQuizz pour les quizz en CM

```python
notesCC=[u'Prasi', u'Yomig', u'Kartin V', u'Kartin T', u'Nemas V', u'Nemas T',
         u'Coupe-Mot', u'Coupe-Morphe'
        ]
notesQuizz=[u'Quizz04', u'Quizz05', u'Quizz06', u'Quizz07', u'Quizz08', u'Quizz09']
for col in notesCC+notesQuizz:
    dfNotes[col]=dfNotes[col].astype(float)
```

```python
dfNotes[[u'Prasi', u'Yomig', u'Kartin V', u'Kartin T', u'Nemas V', u'Nemas T',
         u'Coupe-Mot', u'Coupe-Morphe',
         u'Quizz04', u'Quizz05', u'Quizz06', u'Quizz07', u'Quizz08', u'Quizz09'
        ]]
```

### Calcul des notes avec bonification
Toutes les notes du CC comptent et on inclue aussi toutes les notes des Quizz en partant de la meilleure une par une
- *notes* contient toutes les moyennes calculées
- on retourne le max de *notes*

```python
def bonus(x):
    if x.notnull().any():
        notes=[]
        nbCC=8
        sommeNotes=float(x[notesCC].sum())
        bonusNotes=sorted([n for n in x[notesQuizz].tolist() if not np.isnan(n)],reverse=True)
        nbNotes=nbCC
        notes.append(round(sommeNotes/nbNotes+.25))
#        print "moyenne brut",sommeNotes/nbNotes,"bonus",bonusNotes
        for b in bonusNotes:
            sommeNotes+=b
            nbNotes+=1
            notes.append(round(sommeNotes/nbNotes+.25))
#        print notes
        return max(notes)
    else:
        return np.nan
        
```

### Ajout de la colonne **Notes** à la structure et exportation
- apply(lambda x: bonus(x), axis=1) permet d'appliquer la fonction bonus à chaque ligne
 - la fonction bonus reçoit une ligne de données complète
 - les valeurs sont accessibles par colonne

```python
dfNotes["Note"]=dfNotes[notesCC+notesQuizz].apply(lambda x: bonus(x),axis=1)
```

```python
dfNotes.to_csv(path_or_buf=notesNom.replace(".csv","-Final.csv"),sep="\t",encoding="utf8")
```
