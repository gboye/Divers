```python
import vobject
import codecs,unidecode,re
```

```python
nVCF="/Users/gilles/Downloads/teledir.vcf"
with codecs.open(nVCF, encoding='utf-8') as inFile:
    obj = vobject.readComponents(inFile.read())
contacts = [contact for contact in obj]
```

```python
lines=[]
for c in contacts:
    print (c.fn.value,c.tel.value)
    lines.append(c.serialize())
```

```python
nSupp="Bouhaben Chef Théba_P Immo Mi_Ray Michel_Boy Mimoune_P Vaujany Nicole_Laf Plume64 Régis".split(" ")
tSupp="0148728490"

vContacts=[]
for c in contacts:
    if c.tel.value==tSupp:
        print (c.fn.value,c.tel.value)
        continue
    for supp in nSupp: 
        if supp.replace("_"," ") in c.fn.value:
            print (c.fn.value,c.tel.value)
            break
    else:
        contact=c.serialize().replace(";", "").replace("TEL:", "TEL;HOME:")
        vContacts.append(contact)
# vContacts
```

### Définition du format d'une entrée

```python
record="""BEGIN:VCARD\r
VERSION:2.1\r
FN:%s\r
N:%s\r
TEL;HOME:%s\r
END:VCARD\r\n"""
```

### Insertion de nouveaux numéros

```python
rInserts="Alain Tallis,07 61 59 78 67;André Fournet,0668569905;Buyang,00 60 17 247 7746;Calimero,0660807555;DeBDX,0670520358;Evelyne Pons,0689856056;Mizaga,0612862115;Jacqueline Lacoste,0671676198;Mo13Luis,0034661673359;Najat Bolivar,00447888525264;Crispol40 P,0638717275;Crispol40,0558890613".split(";")
pInserts=[(nom.strip(),tel.replace(" ","")) for nom,tel in [r.split(",") for r in rInserts]]
pInserts
```

```python
for nom,tel in pInserts:
    contact=record%(nom,nom,tel)
    vContacts.append(contact)
```

### Exportation des VCFs modifiés
- ascii
- utf8 (téléphones noirs)
- latin1 (téléphone blanc)

```python
with codecs.open(nVCF.replace(".vcf","-ascii.vcf"), "w", encoding='utf-8') as outFile:
    for vContact in sorted(vContacts):
        outFile.write(unidecode.unidecode(vContact)+"\r\n")
```

```python
with codecs.open(nVCF.replace(".vcf","-utf8.vcf"), "w", encoding='utf-8') as outFile:
    for vContact in sorted(vContacts):
        outFile.write(vContact+"\r\n")
```

```python
with codecs.open(nVCF.replace(".vcf","-latin1.vcf"), "w", encoding='latin1') as outFile:
    for vContact in sorted(vContacts):
        outFile.write(vContact+"\r\n")
```

```python
nVCF="/Users/gilles/Downloads/Base.vcf"
with codecs.open(nVCF, encoding='utf-8') as inFile:
    lines=inFile.readlines()
lines
```

```python
outLines=[]
for line in lines:
    if line.startswith("TEL;"):
        m=re.match(r"^\D*0[67]\d{8}\d*\s*",line)
        if m:
            outLines.append(line.replace("HOME","CELL"))
        else:
            outLines.append(line)
    else:
        outLines.append(line)
```

```python
with codecs.open(nVCF.replace(".vcf","-mod.vcf"),"w",encoding='utf-8') as outFile:
    for line in outLines:
        outFile.write(line)
```
