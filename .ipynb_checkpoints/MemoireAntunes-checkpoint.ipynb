```python
import pandas as pd
import matplotlib.pyplot as plt
from unidecode import unidecode
```

```python
xls = pd.ExcelFile('/Users/gilles/Github/Divers/Antunes/M2-Antunes.xlsx')

# Now you can list all sheets in the file
xls.sheet_names
# ['house', 'house_extra', ...]
```

```python
# to read just one sheet to dataframe:
df = pd.read_excel(xls, sheet_name=u'Corpus compos\xe9s avec \u306e ')
```

```python
df.columns=[unidecode(col) for col in df.columns.tolist()]
```

```python
pC1=u"position Noyau C1"
pC2=u"position Noyau C2"
df[pC1]=df[pC1].astype(unicode).apply(unidecode)
df[pC2]=df[pC2].astype(unicode).apply(unidecode)
```

```python
df[df[pC1]!=u"???"].groupby(pC1)[[u"Lexeme"]].count().sort_values(u"Lexeme").plot.pie(u"Lexeme",figsize=(10,10),title="C1",cmap='RdYlGn_r')
df[df[pC2]!=u"???"].groupby(pC2)[[u"Lexeme"]].count().sort_values(u"Lexeme").plot.pie(u"Lexeme",figsize=(10,10),title="C2",cmap='RdYlGn_r')
```
