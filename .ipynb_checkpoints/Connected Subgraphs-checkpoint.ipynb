### Connected subgraphs

https://stackoverflow.com/questions/54440779/how-to-find-all-connected-subgraph-of-a-graph-in-networkx

```python
import networkx as nx
import itertools as it
import datetime
```

```python
piba=nx.Graph()
piba.add_edge("S","S2")
piba.add_edge("S","S3")
piba.add_edge("S","S4")
piba.add_edge("S","S5")
piba.add_edge("S","S6")
piba.add_edge("S","S7")
piba.add_edge("S","S8")

piba.add_edge("S2","S3")
piba.add_edge("S2","S4")
piba.add_edge("S5","S7")
piba.add_edge("S6","S8")
```

```python
G = piba.copy()
all_connected_subgraphs = []
               
for nb_nodes in range(2, G.number_of_nodes()):
    for SG in (G.subgraph(selected_nodes) for selected_nodes in it.combinations(G, nb_nodes)):
        if nx.is_connected(SG):
            print(", ".join(sorted(SG.nodes)))
            all_connected_subgraphs.append(SG.nodes)
print len(all_connected_subgraphs)
```

```python
connect2=0
for sg1, sg2 in it.combinations(all_connected_subgraphs,2):
    i=set(sg1).intersection(set(sg2))
    if len(i)==0:
        connect2+=1
        print sg1, sg2
print connect2
```

```python
connect3=0
for sg1, sg2, sg3 in it.combinations(all_connected_subgraphs,3):
    i=len(set(sg1).intersection(set(sg2)))+\
    len(set(sg2).intersection(set(sg3)))+\
    len(set(sg1).intersection(set(sg3)))
    if i==0:
        connect3+=1
        print sg1, sg2,sg3
print connect3
```
