# Get authors from .bib

```python
bibNoms=glob.glob(home+"*.bib")
with open(bibNoms[0]) as input:
    parser = bibtexparser.bparser.BibTexParser()
    parser.customization = bibtexparser.customization.convert_to_unicode
    bibData = bibtexparser.load(input, parser=parser)

auteurSet=set()
exclus=["ar","de","eloise","ito","ra","ra A.","ro","y Allen","'t Hart"]
ajouts=["Trifone"]
for ajout in ajouts:
    auteurSet.add(ajout)
for entry in bibData.entries:
    if "author" in entry:
        auteurs=entry["author"].split("and")
        for auteur in auteurs:
            element=auteur.split(",")[0].strip().replace("{","").replace("}","").replace("~"," ")
            if len(element)>1 and not element in exclus:
                auteurSet.add(element)
    else:
        print "no author"
        print entry
listeAuteurs=sorted(list(auteurSet))
for auteur in listeAuteurs:
    print auteur
```

# Find citations from authors and format

```python
def par(liste):
    return "(%s)"%"|".join(liste)

separateurs=[ur'\s+et',ur'\s+and',"\s+\\\\&",","]
regAuteurs=par(listeAuteurs)
regSeparateurs=par(separateurs)
#print regAuteurs, regSeparateurs
```

```python
def citations(chaine):
    textCitations=[]
    text=chaine
    mCite=re.findall(ur"(%s(%s %s)?('s)? \(\d\d\d\d[^)]*\))"%(regAuteurs,regSeparateurs,regAuteurs),chaine)
    if mCite:
        for e in mCite:
            textCitations.append(e[0])
            mSimple=re.match(ur"%s \((\d\d\d\d)\)"%regAuteurs,e[0])
            if mSimple:
                print "simpleT", mSimple.group(1), mSimple.group(2)
                text=text.replace(mSimple.group(0),ur"\\citet{%s%s}"%(mSimple.group(1),mSimple.group(2)))
            mDouble=re.match(ur"%s%s %s \((\d\d\d\d)\)"%(regAuteurs,regSeparateurs,regAuteurs),e[0])
            if mDouble:
                print "doubleT", mDouble.group(1), mDouble.group(3), mDouble.group(4)
                text=text.replace(mDouble.group(0),ur"\\citet{%s%s%s}"%(mDouble.group(1), mDouble.group(3), mDouble.group(4)))
    m=re.findall(ur"(\([^)]*%s(%s %s)? \d\d\d\d[^)]*\))"%(regAuteurs,regSeparateurs,regAuteurs),chaine)
    if m:
        for e in m:
            textCitations.append(e[0])
            mSimple=re.match(ur"\(%s (\d\d\d\d)\)"%regAuteurs,e[0])
            if mSimple:
                print "simpleP", mSimple.group(1), mSimple.group(2)
                text=text.replace(mSimple.group(0),ur"\\citep{%s%s}"%(mSimple.group(1),mSimple.group(2)))
            mDouble=re.match(ur"\(%s%s %s (\d\d\d\d)\)"%(regAuteurs,regSeparateurs,regAuteurs),e[0])
            if mDouble:
                print "doubleT", mDouble.group(1), mDouble.group(3), mDouble.group(4)
                text=text.replace(mDouble.group(0),ur"\\citep{%s%s%s}"%(mDouble.group(1), mDouble.group(3), mDouble.group(4)))
    print textCitations
    return text
```

```python
ligne=ur"Fradin et Kerleroux (2003), ou Melloni \& Thornton (1990) et Spencer (1990)"
citations(ligne)
```

# Process .tex files

```python
texNoms=glob.glob(home+"*.tex")
texNoms
```

```python
for texNom in texNoms:
    with codecs.open(texNom,"r",encoding="utf8") as input:
        textLines=input.readlines()

    textLines=[l.strip() for l in textLines]

    texLines=[]
    accumulateur=[]
    for textLine in textLines:
        if textLine=="" and accumulateur:
            texLines.append(" ".join(accumulateur))
            accumulateur=[]
        else:
            accumulateur.append(textLine)
    texLines

    for texLine in texLines:
    #    print texLine
        print citations(texLine)
```
