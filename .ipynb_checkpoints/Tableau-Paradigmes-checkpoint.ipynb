```python
def getParadigmes(indexParadigmes):
    df=pd.read_csv(repParadigmes+fParadigmes[indexParadigmes],sep=";",index_col=0,encoding="utf8")
    df.set_index("lexeme",inplace=True)
    paradigmeCases=df.columns.tolist()
    return df, paradigmeCases
```

```python
def paradigme2(lexeme):
    return df.loc[lexeme].dropna().to_dict()
```

```python
def tableau(lexeme):
    lEmptyTableau=True
    lTableau=paradigme1
    lParadigme=paradigme2(lexeme)
    for c in paradigmeCases:
        formeC=""
        if c in lParadigme:
            formeC=lParadigme[c]
            lEmptyTableau=False
        lTableau=lTableau.replace(c,formeC)
    if lEmptyTableau:
        print u"pas de données"
    return lTableau
```
