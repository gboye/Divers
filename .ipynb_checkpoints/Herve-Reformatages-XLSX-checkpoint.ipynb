# Fusion et reformatage de 2 tableaux xlsx dans un template

```python
import pandas as pd
import numpy as np
```

## Lecture des tableaux

```python
df1=pd.read_excel("Tableau1.xlsx")
df2=pd.read_excel("Tableau2.xlsx")
dfTemplate=pd.read_excel("Template.xlsx")
```

## Fusion

### Sélection des données manquantes dans le premier tableau

```python
couplesSerieUpdate=df1.groupby("Engine.SerialNo")["LastUpdate"].first().to_dict()
```

### Fusion avec le deuxième tableau

```python
def lastUpdate(row):
#     print row["Engine S/N"]
    nSerie=row["Engine S/N"]
    row["LastUpdate"]=couplesSerieUpdate[nSerie].strftime("%d/%m/%Y")
    return row
```

```python
df21=df2.apply(lambda x: lastUpdate(x),axis=1)
```

```python
df21[df21["Engine S/N"]==32233]
```

```python
df21.columns
```

```python
df21.rename({'Module 02 Remaining hours to TBO': 'Module 02\nRemaining hours to TBO', 'Contract Type': 'Contract type','LastUpdate':"Update date"}, axis=1,inplace=True)
df21
```

```python
df21.loc[df21.Customer.isnull(),"Customer"]=df21.loc[df21.Customer.isnull(),"Owner"]
df21
```

```python
dfTemplate.drop([0],inplace=True)
```

```python
colonnesCommunes=dfTemplate.columns.to_list()[:10]+["Update date"]
```

```python
dfTemplate[colonnesCommunes]=df21[colonnesCommunes]
```

```python
dfTemplate.drop_duplicates(keep="first", inplace=True)
```

```python
dfTemplate#[dfTemplate["Engine S/N"]==32382]
```

```python
dfTransfert=df21.groupby(["Engine S/N", "LifeLimitedParts.Type", "Counter"])[["PartNo","LifeLimitedParts.SerialNo","Consumed"]].first().reset_index()
```

```python
dfPTDisc=dfTransfert[dfTransfert["LifeLimitedParts.Type"]=="PT Disc"]
dfPTDisc
```

```python
def addPTDisc(row):
    nSerie=row["Engine S/N"]
    if len(dfPTDisc.loc[dfPTDisc["Engine S/N"]==nSerie,"LifeLimitedParts.SerialNo"])>0:
#         print nSerie,dfPTDisc.loc[dfPTDisc["Engine S/N"]==nSerie,"LifeLimitedParts.SerialNo"].values[0]
        row["PT Disk S/N"]=dfPTDisc.loc[dfPTDisc["Engine S/N"]==nSerie,"LifeLimitedParts.SerialNo"].values[0]
        row["PT Disk Cycles"]=dfPTDisc.loc[dfPTDisc["Engine S/N"]==nSerie,"Consumed"].values[0]
    return row
```

```python
dfTemplatePTD=dfTemplate.apply(lambda x: addPTDisc(x),axis=1)
dfTemplatePTD
```

```python
dfPTBladesN2=dfTransfert[(dfTransfert["LifeLimitedParts.Type"]=="PT Blades")&(dfTransfert["Counter"]=="N2")]
dfPTBladesN2
```

```python
def addPTBladesN2(row):
    nSerie=row["Engine S/N"]
    if len(dfPTBladesN2.loc[dfPTBladesN2["Engine S/N"]==nSerie,"PartNo"])>0:
        row["PT Blade P/N"]=dfPTBladesN2.loc[dfPTBladesN2["Engine S/N"]==nSerie,"PartNo"].values[0]
    if len(dfPTBladesN2.loc[dfPTBladesN2["Engine S/N"]==nSerie,"Consumed"])>0:
        row["PT Blades Cycles"]=dfPTBladesN2.loc[dfPTBladesN2["Engine S/N"]==nSerie,"Consumed"].values[0]
    return row
```

```python
# dfTemplatePTD.apply(lambda x: addPTBladesN2(x),axis=1)
dfTemplatePTB=dfTemplatePTD.apply(lambda x: addPTBladesN2(x),axis=1)
```

```python
dfPTBladesTSN=dfTransfert[(dfTransfert["LifeLimitedParts.Type"]=="PT Blades")&(dfTransfert["Counter"]=="TSN")]
```

```python
dfPTBladesTSN[dfPTBladesTSN["Engine S/N"]==32030]
```

```python
def addPTBladesTSN(row):
    nSerie=row["Engine S/N"]
    nPartNo=row["PT Blade P/N"]
    if len(dfPTBladesTSN.loc[dfPTBladesTSN["Engine S/N"]==nSerie,"PartNo"])>0:
        row["PT Blade P/N"]=dfPTBladesTSN.loc[dfPTBladesTSN["Engine S/N"]==nSerie,"PartNo"].values[0]
#             print nSerie,nPartNo,row["PT Blade P/N"]
    if len(dfPTBladesTSN.loc[dfPTBladesTSN["Engine S/N"]==nSerie,"Consumed"])>0:
        row["PT Blades TSN"]=dfPTBladesTSN.loc[dfPTBladesTSN["Engine S/N"]==nSerie,"Consumed"].values[0]
    return row
```

```python
dfTemplatePTBTSN=dfTemplatePTB.apply(lambda x: addPTBladesTSN(x),axis=1)
```

```python
SNBlanks=dfTemplatePTBTSN[dfTemplatePTBTSN["PT Blade P/N"].str.contains("BLANK")]["Engine S/N"].to_list()
```

```python
df21[(df21["Engine S/N"].isin(SNBlanks))&(df21["LifeLimitedParts.Type"]=="PT Blades")]
```
