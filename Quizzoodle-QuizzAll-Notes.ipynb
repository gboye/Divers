```python
def assemblerNoms(nom,sep=";"):
    noms=nom.replace(" ",sep).replace("-",sep).replace("'",sep).split(sep)
    result=[]
    for ordre in it.permutations(noms):
        result.append("".join(ordre))
    return result
```

```python
def chunks(l, n):
    """Yield successive n-sized chunks from l."""
    for i in range(0, len(l), n):
        yield l[i:i + n]    
```

## Bonnes réponses pour les différents quizz
- sur la deuxième ligne lQuizz[1], il y a toutes les bonnes réponses
- on coupe le Quizz All par groupe de trois questions pour chaque quizz d'origine
 - *quizzReponses*

```python
def getReponses(ligne):
    result={}
    ligneEls=[l.strip() for l in ligne.split(";")]
    pseudo=ligneEls[0]
    reponses=ligneEls[1:]
    for num,rep in enumerate(chunks(reponses,3)):
        result["%02d"%(num+1)]=rep
    return pseudo,result    
```

```python
def markReponses(reponses):
    scores={}
    for q in quizzReponses:
        score=0
        for n,r in enumerate(quizzReponses[q]):
            if r==reponses[q][n]:
                score+=1
        scores[q]=int(round(float(score)*20/3,0))    
    return scores
```

```python
def getNotes(ligne):
    ligneEls=ligne.split(";")
    etudiant=ligneEls[1]+"\t"+ligneEls[2]
    notes={}
    for num,note in enumerate(ligneEls[6:16]):
        if note=="-": note=0
        notes["%02d"%(num+1)]=int(note)
    return etudiant,notes
```

```python
'''https://stackoverflow.com/a/35491326/3986209'''
def invert_dict(d): 
    inverse = dict() 
    for key in d: 
        # Go through the list that is saved in the dict:
        for item in d[key]:
            # Check if in the inverted dict the key exists
            if item not in inverse: 
                # If not create a new list
                inverse[item] = [key] 
            else: 
                inverse[item].append(key) 
    return inverse
```
